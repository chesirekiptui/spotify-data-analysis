```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the tracks dataset
tracks = pd.read_csv('C:/Users/user/Desktop/Data Analysis Projects/Spotify Data Analysis/tracks.csv')
tracks.head()
```

```python
#Checking for null values
pd.isnull(tracks).sum()
```

```python
tracks.info()
```

```python
#Finding the top 10 least popular songs

least_popular = tracks.sort_values('popularity', ascending = True).head(10)
least_popular
```

```python
tracks.describe().transpose()
```

```python
most_popular = tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
#Formatting the release_date column to datetime
tracks.set_index("release_date", inplace = True)
tracks.index = pd.to_datetime(tracks.index)
tracks.head()
```

```python
# Finding the name of the artist who is in position 1000
tracks[["artists"]].iloc[1000]
```

```python
tracks['duration'] = tracks['duration_ms'].apply(lambda x : round(x/1000))
tracks.drop('duration_ms', inplace=True, axis=1)
```

```python
tracks.duration.head()
```

```python
#Visualizing the data using a heatmap

corr_tracks=tracks.drop(['key','mode','explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14,8))
heatmap=sns.heatmap(corr_tracks, annot=True, fmt='.1g')
heatmap.set_title('Correlation Heatmap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# Getting a sample of the data
sample_tracks = tracks.sample(int(0.005*len(tracks)))
print(len(sample_tracks))
```

```python
# Loudness Vs Energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_tracks, y='loudness', x='energy', color='b').set(title='Loudness Vs Energy')
```

```python
# Popularity Vs Acousticness
plt.figure(figsize=(10,6))
sns.regplot(data=sample_tracks, y='popularity', x='acousticness', color='g').set(title='Popularity Vs Acousticness')
```

```python
tracks['dates'] = tracks.index.get_level_values('release_date')
tracks.dates = pd.to_datetime(tracks.dates)
years = tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True , aspect=2, height=5 , kind='hist').set(title='Number Of Song Per Years')
```

```python
total_duration = tracks.duration
fig_dims=(20,8)
fig, ax =plt.subplots(figsize= fig_dims)
fig=sns.barplot(x=years,y=total_duration, ax=ax,errwidth=False).set(title='Year Vs Duration')
plt.xticks(rotation=90)
```

```python
sns.set_style(style='whitegrid')
fig_dims=(10,5)
fig, ax =plt.subplots(figsize= fig_dims)
fig=sns.lineplot(x=years,y=total_duration, ax=ax).set(title='Year Vs Duration')
plt.xticks(rotation=60)
```
